```python
import pandas as pd
import numpy as np
```

```python
abiotic_df = pd.read_excel('../data/ABIO.xlsx', sheet_name='ABIO_COMBINED')
biotic_df = pd.read_excel('../data/PHYTO.xlsx', sheet_name='PHYTO_FLATTENED')

display(abiotic_df)

display(biotic_df)
```

```python
abiotic_df["DATUMTIJDWAARDE"] = pd.to_datetime(abiotic_df["DATUMTIJDWAARDE"], format='mixed')
abiotic_df["DATUMTIJDWAARDE"] = abiotic_df["DATUMTIJDWAARDE"].dt.date
abiotic_df
```

```python
biotic_df["DATUMTIJDWAARDE"] = pd.to_datetime(biotic_df["DATUMTIJDWAARDE"], format='mixed')
biotic_df["DATUMTIJDWAARDE"] = biotic_df["DATUMTIJDWAARDE"].dt.date
biotic_df
```

```python
abiotic_measurements = list(zip(abiotic_df["LOC_CODE"], abiotic_df["DATUMTIJDWAARDE"]))
biotic_measurements = list(zip(biotic_df["LOC_CODE"], biotic_df["DATUMTIJDWAARDE"]))


similarity = []
for abiotic in abiotic_measurements:
    for biotic in biotic_measurements:
        if abiotic[0] == biotic[0]:  # Check if locations are the same
            if abs((abiotic[1] - biotic[1]).days) <= 1:  # Check if dates are within one week
                similarity.append(abiotic)
                break

print(" abiotic measurements = ", len(abiotic_measurements), '\n', 
      "biotic measurements = ", len(biotic_measurements), '\n',
      "similarity = ", len(similarity))

similarity
```

```python
dups = abiotic_df.loc[abiotic_df[["LOC_CODE", "DATUMTIJDWAARDE"]].duplicated(keep=False)]
```

```python
dups = dups.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])
display(dups)
```

```python


def custom_aggregate(series):
    if series.notna().any():
        return series.mean()
    return None

# Identify the columns that need aggregation
agg_columns = abiotic_df.columns.difference(['LOC_CODE', 'DATUMTIJDWAARDE'])

# Create the aggregation dictionary
agg_functions = {col: custom_aggregate for col in agg_columns}

abiotic_agregated = dups.groupby(["LOC_CODE", "DATUMTIJDWAARDE"]).agg(agg_functions).reset_index()

abiotic_agregated["AGG"] = True

```

```python
abiotic_deduped = abiotic_df.drop_duplicates(subset=["LOC_CODE", "DATUMTIJDWAARDE"], keep=False)
abiotic_deduped["AGG"] = False

abiotic_updated = pd.concat([abiotic_deduped, abiotic_agregated], ignore_index=True)

abiotic_updated
```

### Check if duplicates are now gone

```python
abiotic_updated.loc[abiotic_updated[["LOC_CODE", "DATUMTIJDWAARDE"]].duplicated(keep=False)].sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])
```

### Now perform all this on the phyto data


```python
abiotic_updated = abiotic_updated.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])

filename = '../data/ABIO.xlsx'
sheetname = 'ABIO_AGGREGATED'

with pd.ExcelWriter(filename, mode='a') as writer:  
    abiotic_updated.to_excel(writer, sheet_name=sheetname, index=False)

print(f"Data written to sheet '{sheetname}' in '{filename}'")
```

```python
biotic_df = biotic_df.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])

filename = '../data/PHYTO.xlsx'
sheetname = 'PHYTO_ONDATE'

with pd.ExcelWriter(filename, mode='a') as writer:  
    biotic_df.to_excel(writer, sheet_name=sheetname, index=False)

print(f"Data written to sheet '{sheetname}' in '{filename}'")
```
