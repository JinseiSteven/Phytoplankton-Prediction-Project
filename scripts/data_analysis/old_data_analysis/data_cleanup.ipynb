```python
import pandas as pd
import numpy as np
```

```python
abiotic_df = pd.read_excel('../data/ABIO.xlsx', sheet_name='ABIO_COMBINED')
biotic_df = pd.read_excel('../data/PHYTO.xlsx', sheet_name='PHYTO_FLATTENED')

display(abiotic_df)

display(biotic_df)
```

```python
abiotic_df["DATUMTIJDWAARDE"] = pd.to_datetime(abiotic_df["DATUMTIJDWAARDE"], format='%d-%m-%Y')
abiotic_df["DATUMTIJDWAARDE"] = abiotic_df["DATUMTIJDWAARDE"].dt.date
abiotic_df
```

```python
biotic_df["DATUMTIJDWAARDE"] = pd.to_datetime(biotic_df["DATUMTIJDWAARDE"], format='mixed')
biotic_df["DATUMTIJDWAARDE"] = biotic_df["DATUMTIJDWAARDE"].dt.date
biotic_df
```

```python
abiotic_measurements = list(zip(abiotic_df["LOC_CODE"], abiotic_df["DATUMTIJDWAARDE"]))
biotic_measurements = list(zip(biotic_df["LOC_CODE"], biotic_df["DATUMTIJDWAARDE"]))


similarity = []
for abiotic in abiotic_measurements:
    for biotic in biotic_measurements:
        if abiotic[0] == biotic[0]:  # Check if locations are the same
            if abs((abiotic[1] - biotic[1]).days) <= 1:  # Check if dates are within one week
                similarity.append(abiotic)
                break

print(" abiotic measurements = ", len(abiotic_measurements), '\n', 
      "biotic measurements = ", len(biotic_measurements), '\n',
      "similarity = ", len(similarity))

similarity
```

```python
dups = abiotic_df.loc[abiotic_df[["LOC_CODE", "DATUMTIJDWAARDE"]].duplicated(keep=False)]
```

```python
dups = dups.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])
display(dups)
```

```python


def custom_aggregate(series):
    if series.notna().any():
        return series.mean()
    return None

# Identify the columns that need aggregation
agg_columns = abiotic_df.columns.difference(['LOC_CODE', 'DATUMTIJDWAARDE'])

# Create the aggregation dictionary
agg_functions = {col: custom_aggregate for col in agg_columns}

abiotic_agregated = dups.groupby(["LOC_CODE", "DATUMTIJDWAARDE"]).agg(agg_functions).reset_index()

abiotic_agregated["AGG"] = True

```

```python
abiotic_deduped = abiotic_df.drop_duplicates(subset=["LOC_CODE", "DATUMTIJDWAARDE"], keep=False)
abiotic_deduped["AGG"] = False

abiotic_updated = pd.concat([abiotic_deduped, abiotic_agregated], ignore_index=True)

abiotic_updated
```

### Merge datasets on time TODO

```python
def merge_datasets(df1, df2, time_column, location_column):
    """
    Merges two datasets on time and location. 
    Dataset df2 is interpolated to match the nearest times of df1.
    
    Parameters:
    - df1 (pd.DataFrame): The first dataset with higher frequency.
    - df2 (pd.DataFrame): The second dataset to be interpolated.
    - time_column (str): The name of the time column.
    - location_column (str): The name of the location column.
    
    Returns:
    - pd.DataFrame: The merged dataset.
    """
    df1[time_column] = pd.to_datetime(df1[time_column])
    df2[time_column] = pd.to_datetime(df2[time_column])

    # Ensure the datasets are sorted by time and location
    df1 = df1.sort_values(by=[time_column, location_column])
    df2 = df2.sort_values(by=[time_column, location_column])
    
    # Merge datasets using the nearest previous time
    merged_df = pd.merge_asof(df1, df2, on=time_column, by=location_column, direction='backward', suffixes=('_df1', '_df2'))
    
    return merged_df

abiotic_updated = abiotic_updated.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])
biotic_df = biotic_df.sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])

merged_dataset = merge_datasets(abiotic_updated, biotic_df, 'DATUMTIJDWAARDE', 'LOC_CODE').sort_values(by=["LOC_CODE", "DATUMTIJDWAARDE"])
```

```python
irradiance_df = irradiance_df.rename(columns={'YYYYMMDD': 'DATUMTIJDWAARDE'})
display(irradiance_df.head())
```

```python
print(irradiance_df.columns)
irradiance_updated = irradiance_df[['DATUMTIJDWAARDE', '    Q', 'PAR [J/m2d]', 'PAR [kJ/m2d]', 'kPAR_7d', 'kPAR_14d']]
display(irradiance_updated.head())
```

### Merge Merged dataset with Irradiance data

```python
# first sort both dataset to time

mergable_merged = merged_dataset.sort_values(by='DATUMTIJDWAARDE')
mergable_irradiance = irradiance_updated.sort_values(by='DATUMTIJDWAARDE')

# then convert to same value

mergable_merged['DATUMTIJDWAARDE'] = pd.to_datetime(mergable_merged['DATUMTIJDWAARDE'])
mergable_irradiance['DATUMTIJDWAARDE'] = pd.to_datetime(mergable_irradiance['DATUMTIJDWAARDE'])


merged_updated = pd.merge_asof(mergable_merged, mergable_irradiance, on='DATUMTIJDWAARDE', direction='nearest').sort_values(by=['LOC_CODE', 'DATUMTIJDWAARDE'])
display(merged_updated)
```

### Write to excel file

```python
filename = '../data/MERGED_DATA.xlsx'
sheetname = 'AGGREGATED_MERGE'

with pd.ExcelWriter(filename, mode='w') as writer:  
    merged_updated.to_excel(writer, sheet_name=sheetname, index=False)

print(f"Data written to sheet '{sheetname}' in '{filename}'")
```
