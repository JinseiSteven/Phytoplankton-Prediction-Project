```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score
from fastdtw import fastdtw
from sklearn.cluster import KMeans
```

```python
# Create a df containing only biotic data
df = pd.read_excel('../data/MERGED_DATA_180624.xlsx', sheet_name='MERGE_FINAL')
cols = df.columns.tolist()
# Columns with the phytoplankton
subset = cols[26:-1]
df_bio = df[subset]
# Add date and time for time series
# df_bio.insert(0, "Datum", df['DATUM'], True)
# df_bio.insert(0, "Tijd", df['TIJD'], True)
df_bio.head()
```

```python
# Drop columns containing only nans, if any
df_bio.dropna(axis=1, how='all', inplace=True)
df_bio.dropna(how='all', inplace=True)

# Fill remaining nans with zeros
df_bio.fillna(0, inplace=True)

# Convert dataframe to a numpy array
data = df_bio.to_numpy().T

similarity_matrix = np.zeros((len(data), len(data)))

distance_matrices = []
# checking similarity for all series pairs
for i, series1 in enumerate(data):
    for j, series2 in enumerate(data):
        distance, path = fastdtw(series1, series2)
        similarity_matrix[i, j] = distance
        print(f"Pair ({i}, {j}) done - {int((i * len(data) + j) / ((len(data))**2) * 100)}%")

data.shape
```

```python
n_clusters = 3
model = KMeans(n_clusters=n_clusters)
preds = model.fit_predict(data)
print(preds)
```

```python
for i in range(20):
    model = KMeans(n_clusters=n_clusters)
    preds = model.fit_predict(data)
    score = silhouette_score(similarity_matrix, preds)
    print(f"idx: {i}, s: {score}")
```
